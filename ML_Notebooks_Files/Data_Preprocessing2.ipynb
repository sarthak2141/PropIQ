```python
import re

def normalize(text):
    text = str(text).lower()

    # remove duplicate words: "road road"
    text = re.sub(r'\b(\w+)\s+\1\b', r'\1', text)

    # remove punctuation
    text = re.sub(r'[^a-z0-9\s]', ' ', text)

    # remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text

df["sector"] = df["sector"].apply(normalize)
```

```python
def fill_sector(text):
    if pd.isna(text):
        return text
    text_l=text.lower().strip()
    if 'sector' in text_l:
        return text_l
    if text_l in sector_map.keys():
        return 'sector '+ str(sector_map[text])
    return text_l
```
